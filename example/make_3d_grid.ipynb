```python
from pyionoseis.source import EarthquakeSource
source = EarthquakeSource(r"event.toml")
print(source)
```

```python
from pyionoseis.model import Model3D
model = Model3D(r"event.toml")
print(model)
```

```python
model.assign_source(source)
model.make_3Dgrid()
model.assign_atmosphere()
print(model)
```

```python
model.plot_source()
# make plot grid
model.plot_grid(show_gridlines=True)
model.plot_grid_3d()

model.atmosphere.plot()
```
